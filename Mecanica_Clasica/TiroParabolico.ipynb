# Tiro Parab√≥lico (Newton)

```python
%reset -f

import sympy as smp
import numpy as np

from sympy.solvers.ode.systems import dsolve_system
from scipy.integrate import solve_ivp

import matplotlib.pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
```

Define los simbolos apropiados en sympy

```python
t, g, m = smp.symbols('t g m', positive=True)
Fx, ax, Fy, ay = smp.symbols(r'F_x a_x F_y a_y', real=True)
```

$$\vec{F} = m\vec{a}$$

Al ser tiro parab√≥lico tenemos que

$$F_{y} = ma_{y}$$

```python
Eqy = smp.Eq(Fy, m*ay)
Eqy
```

Y que

$$F_{x} = ma_{x}$$

```python
Eqx = smp.Eq(Fx, m*ax)
Eqx
```

$y$ y $x$ son dependientes del tiempo

```python
y = smp.Function('y', real=True)
y = y(t)

x = smp.Function('x', real=True)
x = x(t)
```

Definimos las dos primeras derivadas temporales de $y$ e  $x$

```python
d1_y = smp.diff(y, t)
d2_y = smp.diff(d1_y, t)

d1_x = smp.diff(x, t)
d2_x = smp.diff(d1_x, t)
```

La fuerza de gravedad solo act√∫a en el eje $y$

```python
F_y = -m*g
F_y
```

La acelerac√≠on en $y$ es: $a_y = \ddot{y}$ <br>
Se sustituyen fuerzas y aceleraciones en la ecuaci√≥n para $y$:

```python
Eqy = Eqy.subs([
    [Fy, F_y],
    [ay, d2_y]
])
Eqy
```

La acelerac√≠on en $x$ es: $a_x = \ddot{x}$ <br>
No hay ninguna fuerza actuando en $x$, por lo que: $F_x = 0$ <br>
Se sustituye esto en la ecuaci√≥n para $x$

```python
Eqx = Eqx.subs([
    [Fx, 0],
    [ax, d2_x]
])
Eqx
```

## Soluci√≥n Exacta (integrate) para Ecuaciones Desacopladas

### Soluci√≥n para y

```python
Eqy
```

La ecuaci√≥n de movimiento obtenida para $y$ se puede resolver por separaci√≥n de variables e integrando, primero tenemos que despejar la segunda derivada $\ddot{y}$

```python
d2_y_explicit = smp.solve([Eqy], [d2_y])[d2_y]
d2_y_explicit
```

Definimos dos constantes (de integraci√≥n) $C_1$ y $C_2$ e integramos 2 veces con respecto a $t$ para obtener la funci√≥n de velocidad y posici√≥n en $y$

```python
C1, C2 = smp.symbols('C1 C2')
```

```python
d1_y_explicit = smp.integrate(d2_y_explicit, t) + C2
d1_y_explicit
```

```python
y_explicit = smp.integrate(d1_y_explicit, t) + C1
y_explicit
```

Integrando dos veces, se obtuvo la soluci√≥n exacta de la ecuaci√≥n de movimiento para $y$

### Soluci√≥n para x

```python
Eqx
```

La ecuaci√≥n de movimiento obtenida para $x$ se puede resolver por separaci√≥n de variables e integrando, primero tenemos que despejar la segunda derivada $\ddot{x}$

```python
d2_x_explicit = smp.solve([Eqx], [d2_x])[d2_x]
d2_x_explicit
```

Integramos 2 veces con respecto a $t$ para obtener la funci√≥n de velocidad y posici√≥n en $x$

```python
d1_x_explicit = smp.integrate(d2_x_explicit, t) + C2
d1_x_explicit
```

```python
x_explicit = smp.integrate(d1_x_explicit, t) + C1
x_explicit
```

Integrando dos veces, se obtuvo la soluci√≥n exacta de la ecuaci√≥n de movimiento para $x$

## Soluci√≥n Numerica

Volvemos a definir algunos simbolos ya que los hab√≠amos sobre-escrito

```python
t, g, m = smp.symbols('t g m', positive=True)
```

Recordemos el despeje de la segunda derivada de $\ddot{y}$ y $\ddot{x}$

```python
d2_y_explicit
```

```python
d2_x_explicit
```

Para resolver numericamente las ecuaciones de segundo orden se tiene que transformar en un sistema de ecuaciones de primer orden, para eso realizamos la siguiente sustituci√≥n <br>
<ul>
  <li>$\dot{y} = z_1$</li>
  <li>$\dot{z}_1 = \ddot{y} = -g$</li>
  <li>$\dot{x} = z_2$</li>
  <li>$\dot{z}_2 = \ddot{x} = 0$</li>
</ul> 
y este sistema lo transformamos en funciones n√∫mericas

```python
dy_dt_f = smp.lambdify(d1_y, d1_y, modules='numpy')
dz1_dt_f = smp.lambdify((t, g, m, y, d1_y), d2_y_explicit, modules='numpy')

dx_dt_f = smp.lambdify(d1_x, d1_x, modules='numpy')
dz2_dt_f = 0
```

Se define un vector $\vec{S} = (y, z1, x, z2)$ y una funci√≥n que toma $\vec{S}$ y $t$, y devuelve $\dot{S}$ para poder resolver el sistema de ecuaciones

```python
def dSdt(t, S, g, m):
    y, z1, x, z2 = S
    return [
        dy_dt_f(z1),
        dz1_dt_f(t, g, m, y, z1),
        dx_dt_f(z2),
        dz2_dt_f
    ]
```

Definimos un arreglo de puntos en el tiempo en el cual queremos obtener la soluci√≥n

```python
t0 = 0
tf = 10
t_size = 1001
t = np.linspace(t0, tf, t_size)
```

Se definen algunas constantes (aceleraci√≥n de la gravedad y masa) que se guardan en la lista args y se definen las condiciones iniciales (posici√≥n-y:y0, velocidad-y: vy0, posici√≥n-x:x0, velocidad-x: vx0) que se guardan en la lista S0

```python
g = 9.81
m = 2
args = [g, m]
y0 = 0
vy0 = 52
x0 = 0
vx0 = 2
S0 = [y0, vy0, x0, vx0]
```

Se utiliza la funci√≥n solve_ivp para resolver la ecuaci√≥n diferencial

```python
sol = solve_ivp(dSdt, t_span=(t0, tf), y0=S0, method='RK45', t_eval=t, args=args)
```

Guardamos la soluci√≥n y se grafica

```python
y_sol_numeric = sol.y[0]
d1_y_sol_numeric = sol.y[1]

x_sol_numeric = sol.y[2]
d1_x_sol_numeric = sol.y[3]
```

```python
plt.figure(figsize=(8, 8))
plt.plot(t, y_sol_numeric, lw=3, c='blue', label='y-position')
plt.plot(t, d1_y_sol_numeric, lw=3, c='red', label='y-speed')
plt.plot(t, x_sol_numeric, lw=3, c='orange', label='x-position')
plt.plot(t, d1_x_sol_numeric, lw=3, c='black', label='x-speed')
plt.xlabel('time (s)')
plt.ylabel('distance (m)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.plot(y_sol_numeric, d1_y_sol_numeric, lw=3, c='blue', label='y-phase')
plt.plot(x_sol_numeric, d1_x_sol_numeric, lw=3, c='red', label='x-phase')
plt.xlabel('distance (m)')
plt.ylabel('speed (m/s)')
plt.legend()
plt.show()
```
