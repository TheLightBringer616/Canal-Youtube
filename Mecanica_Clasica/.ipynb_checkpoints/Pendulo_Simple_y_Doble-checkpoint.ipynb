# Pendulo Simple (Lagrangiano)

```python
%reset -f

import numpy as np
import sympy as smp

from scipy.integrate import solve_ivp

import matplotlib.pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
```

Define los simbolos apropiados en sympy

```python
t, g = smp.symbols('t g', real=True, positive=True)
m = smp.symbols('m', real=True, positive=True)
L = smp.symbols('L', real=True, positive=True)
```

Definimos a $\theta$ y a sus dos primeras derivadas temporales

```python
theta = smp.symbols(r'\theta', cls=smp.Function)(t)
d1_theta = smp.diff(theta, t)
d2_theta = smp.diff(d1_theta, t)
```

Parametrizamos el movimiento de la masa: <br> <br>

En **x**: $$x = L\sin(\theta)$$ <br> <br>

En **y**: $$y = -L\cos(\theta)$$

```python
x = L * smp.sin(theta)
y = -L * smp.cos(theta)

vx = smp.diff(x, t)
vy = smp.diff(y, t)

v_squared = (vx**2 + vy**2).simplify(force=True)
```

Transformamos las funciones simbólicas de velocidad y posición a funciones numéricas

```python
x_f = smp.lambdify((L, theta), x, modules='numpy')
y_f = smp.lambdify((L, theta), y, modules='numpy')

vx_f = smp.lambdify((L, theta, d1_theta), vx, modules='numpy')
vy_f = smp.lambdify((L, theta, d1_theta), vy, modules='numpy')
```

Definimos la energía cinética y potencial. <br>
Se construye la función Lagrangiana: $L = T - U$

```python
T = smp.Rational(1, 2) * m * v_squared
U = m*g*y
Lag = T - U
```

Se obtienen las ecuaciones de movimiento:  <br> <br>

$$\dot{p}_{q} - \nabla_{q}L = 0$$ <br> <br>

donde:  <br> <br>

$$p_{q} = \nabla_{\dot{q}}L$$

```python
gen_p = smp.diff(Lag, d1_theta)
gen_f = smp.diff(Lag, theta)
eq = smp.diff(gen_p, t) - gen_f
```

Se despeja de la ecuación de moviemiento $\ddot{\theta}$

```python
d2_theta_explicit = smp.solve(eq, d2_theta, simplify=False, Rational=False)[0]
d2_theta_explicit
```

Para resolver numericamente la ecuación de segundo orden se tiene que transformar en un sistema de ecuaciones de primer orden, para eso realizamos la siguiente sustitución <br>
<ul>
  <li>$\dot{\theta} = z$</li>
  <li>$\dot{z} = \ddot{\theta} = ...$</li>
</ul> 
y este sistema lo transformamos en funciones númericas

```python
dtheta_dt_f = smp.lambdify(d1_theta, d1_theta, modules='numpy')
dz_dt_f = smp.lambdify((t, g, m, L, theta, d1_theta), d2_theta_explicit, modules='numpy')
```

Se define un vector $\vec{S} = (y, z1, x, z2)$ y una función que toma $\vec{S}$ y $t$, y devuelve $\dot{S}$ para poder resolver el sistema de ecuaciones

```python
def dSdt(t, S, g, m, L):
    theta, z = S
    return [
        dtheta_dt_f(z),
        dz_dt_f(t, g, m, L, theta, z)
    ]
```

Definimos un arreglo de puntos en el tiempo en el cual queremos obtener la solución

```python
t0 = 0
tf = 40
t_size = 4001
t = np.linspace(t0, tf, t_size)
```

Se definen algunas constantes (aceleración de la gravedad y masa) que se guardan en la lista args y se definen las condiciones iniciales $\theta_{0}$ y $\dot{\theta}_{0}$ y  que se guardan en la lista S0

```python
g = 9.81
m = 2
L = 3
args = [g, m, L]
S0 = [np.pi/2, 0]
```

Se utiliza la función solve_ivp para resolver la ecuación diferencial

```python
sol = solve_ivp(dSdt, t_span=(0, 40), y0=S0, method='DOP853', t_eval=t, vectorized=False, args=args)
```

Guardamos la solución y se grafica

```python
theta_sol = sol.y[0]
d1_theta_sol = sol.y[1]
```

```python
plt.figure(figsize=(8, 8))

plt.plot(t, theta_sol, lw=3, c='blue', label='angular position')
plt.plot(t, d1_theta_sol, lw=3, c='red', label='angular speed')

plt.xlabel('time (s)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))

plt.plot(theta_sol, d1_theta_sol, lw=3, c='blue', label='phase')

plt.ylabel('angular speed (rad/s)')
plt.xlabel('angular position (rad)')
plt.legend()
plt.show()
```

Creamos funciones donde ingresamos las funciones númericas de posición y velocidad anteriormente transformadas

```python
def eval_positions(L, theta):
    return [
        x_f(L, theta),
        y_f(L, theta)
    ]

def eval_velocities(L, theta, d1_theta):
    return [
        vx_f(L, theta, d1_theta),
        vy_f(L, theta, d1_theta)
    ]
```

Calculamos las posiciones y velocidades en $x$ e $y$

```python
x, y = eval_positions(L, theta_sol)
vx, vy = eval_velocities(L, theta_sol, d1_theta_sol)
```

```python
plt.figure(figsize=(8, 8))

plt.plot(t, x, lw=3, c='blue', label='x-position')
plt.plot(t, y, lw=3, c='red', label='y-position')

plt.xlabel('time (s)')
plt.ylabel('position (m)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))

plt.plot(x, y, lw=3, c='blue', label='xy-position')

plt.xlabel('x-position (m)')
plt.ylabel('y-position (m)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))

plt.plot(t, vx, lw=3, c='blue', label='x-speed')
plt.plot(t, vy, lw=3, c='red', label='y-speed')

plt.xlabel('time (s)')
plt.ylabel('speed (m/s)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))

plt.plot(vx, vy, lw=3, c='blue', label='xy-speed')

plt.xlabel('x-speed (m/s)')
plt.ylabel('y-speed (m/s)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))

plt.plot(t, np.sqrt(vx**2 + vy**2), lw=3, c='blue', label='speed')
plt.plot(t, np.sqrt(x**2 + y**2), lw=3, c='red', label='distance')

plt.xlabel('time (s)')
plt.legend()
plt.show()
```

```python
fps = len(t[t<1])
frames = t_size - 1
interval = (tf/(t_size-1))*1000

def animate(i):
    ln1.set_data([0, x[i]], [0, y[i]])

fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.set_facecolor('k')
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])
ln1, = plt.plot([], [], 'ro--', lw=3, markersize=8)
ax.set_xlim(-L-1, L+1)
ax.set_ylim(-L-1, L+1)
ani = animation.FuncAnimation(fig, animate, frames=frames, interval=interval)
ani.save('simple_pendulum.mp4', writer='ffmpeg', fps=fps)
```

# Double Pendulum

```python
%reset -f
import numpy as np
import sympy as smp
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
```

```python
t, g = smp.symbols('t g', real=True, positive=True)
m1, m2 = smp.symbols(r'm_1 m_2', real=True, positive=True)
L1, L2 = smp.symbols(r'L_1 L_2', real=True, positive=True)
```

```python
theta1, theta2 = smp.symbols(r'\theta_1 \theta_2', cls=smp.Function)
theta1, theta2 = theta1(t), theta2(t)

d1_theta1 = smp.diff(theta1, t)
d2_theta1 = smp.diff(d1_theta1, t)

d1_theta2 = smp.diff(theta2, t)
d2_theta2 = smp.diff(d1_theta2, t)
```

```python
x1 = L1 * smp.sin(theta1)
y1 = -L1 * smp.cos(theta1)

x2 = x1 + (L2 * smp.sin(theta2))
y2 = y1 - (L2 * smp.cos(theta2))

vx1 = smp.diff(x1, t)
vy1 = smp.diff(y1, t)

vx2 = smp.diff(x2, t)
vy2 = smp.diff(y2, t)

v1_squared = (vx1**2 + vy1**2).simplify(force=True)
v2_squared = (vx2**2 + vy2**2).simplify(force=True)
```

```python
x1_f = smp.lambdify((L1, theta1), x1, modules='numpy')
y1_f = smp.lambdify((L1, theta1), y1, modules='numpy')

vx1_f = smp.lambdify((L1, theta1, d1_theta1), vx1, modules='numpy')
vy1_f = smp.lambdify((L1, theta1, d1_theta1), vy1, modules='numpy')

x2_f = smp.lambdify((L1, theta1, L2, theta2), x2, modules='numpy')
y2_f = smp.lambdify((L1, theta1, L2, theta2), y2, modules='numpy')

vx2_f = smp.lambdify((L1, theta1, d1_theta1, L2, theta2, d1_theta2), vx2, modules='numpy')
vy2_f = smp.lambdify((L1, theta1, d1_theta1, L2, theta2, d1_theta2), vy2, modules='numpy')
```

```python
T1 = smp.Rational(1, 2) * m1 * v1_squared
U1 = m1*g*y1
Lag1 = T1 - U1

T2 = smp.Rational(1, 2) * m2 * v2_squared
U2 = m2*g*y2
Lag2 = T2 - U2

Lag = Lag1 + Lag2
```

```python
gen_p1 = smp.diff(Lag, d1_theta1)
gen_f1 = smp.diff(Lag, theta1)
eq1 = (smp.diff(gen_p1, t) - gen_f1).simplify(force=True)

gen_p2 = smp.diff(Lag, d1_theta2)
gen_f2 = smp.diff(Lag, theta2)
eq2 = (smp.diff(gen_p2, t) - gen_f2).simplify(force=True)
```

```python
d2_theta_explicit = smp.solve([eq1, eq2], [d2_theta1, d2_theta2], simplify=False, Rational=False)
d2_theta1_explicit = d2_theta_explicit[d2_theta1].simplify(force=True)
d2_theta2_explicit = d2_theta_explicit[d2_theta2].simplify(force=True)
```

```python
dtheta1_dt_f = smp.lambdify(d1_theta1, d1_theta1, modules='numpy')
dz1_dt_f = smp.lambdify((t, g, m1, L1, theta1, d1_theta1, m2, L2, theta2, d1_theta2), d2_theta1_explicit, modules='numpy')

dtheta2_dt_f = smp.lambdify(d1_theta2, d1_theta2, modules='numpy')
dz2_dt_f = smp.lambdify((t, g, m1, L1, theta1, d1_theta1, m2, L2, theta2, d1_theta2), d2_theta2_explicit, modules='numpy')
```

```python
def dSdt(t, S, g, m1, L1, m2, L2):
    theta1, z1, theta2, z2 = S
    return [
        dtheta1_dt_f(z1),
        dz1_dt_f(t, g, m1, L1, theta1, z1, m2, L2, theta2, z2),
        dtheta2_dt_f(z2),
        dz2_dt_f(t, g, m1, L1, theta1, z1, m2, L2, theta2, z2)
    ]
```

```python
t0 = 0
tf = 40
t_size = 4001
t = np.linspace(t0, tf, t_size)
g = 9.81
m1 = 2
L1 = 3
m2 = 2
L2 = 3
args = [g, m1, L1, m2, L2]
y0 = [np.pi/2, 0, np.pi/2, 0]
sol = solve_ivp(dSdt, t_span=(0, 40), y0=y0, method='DOP853', t_eval=t, vectorized=False, args=args)
```

```python
theta1_sol = sol.y[0]
d1_theta1_sol = sol.y[1]

theta2_sol = sol.y[2]
d1_theta2_sol = sol.y[3]
```

```python
def eval_positions(L1, theta1, L2, theta2):
    return [
        x1_f(L1, theta1),
        y1_f(L1, theta1),
        x2_f(L1, theta1, L2, theta2),
        y2_f(L1, theta1, L2, theta2)
    ]

def eval_velocities(L1, theta1, d1_theta1, L2, theta2, d1_theta2):
    return [
        vx1_f(L1, theta1, d1_theta1),
        vy1_f(L1, theta1, d1_theta1),
        vx2_f(L1, theta1, d1_theta1, L2, theta2, d1_theta2),
        vy2_f(L1, theta1, d1_theta1, L2, theta2, d1_theta2)
    ]
```

```python
x1, y1, x2, y2 = eval_positions(L1, theta1_sol, L2, theta2_sol)
vx1, vy1, vx2, vy2 = eval_velocities(L1, theta1_sol, d1_theta1_sol, L2, theta2_sol, d1_theta2_sol)
```

```python
fps = len(t[t<1])
frames = t_size - 1
interval = (tf/(t_size-1))*1000

def animate(i):
    ln1.set_data([0, x1[i], x2[i]], [0, y1[i], y2[i]])

fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.set_facecolor('k')
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])
ln1, = plt.plot([], [], 'ro--', lw=3, markersize=8)
ax.set_xlim(-L1-L2-1, L1+L2+1)
ax.set_ylim(-L1-L2-1, L1+L2+1)
ani = animation.FuncAnimation(fig, animate, frames=frames, interval=interval)
ani.save('double_pendulum.mp4', writer='ffmpeg', fps=fps)
```
